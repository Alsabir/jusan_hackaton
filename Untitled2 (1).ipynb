```python
frame = np.ones((len(data_fi),7))*np.NaN
frame = pd.DataFrame(frame, index = data_fi.index,  columns = ["sum_depos","ti","to", "tci", "tco", "dif", "ac"])
for i in tqdm(data_fi.index):
    dKZ = data_fi.loc[i,"deposit_sum_KZT"]
    dUSD = data_fi.loc[i,"deposit_sum_USD"]
    dEUR = data_fi.loc[i,"deposit_sum_EUR"]
    
    if str(dKZ)=="nan" or str(dUSD)=="nan" or str(dEUR)=="nan":
        sum_depos = None
    else:
        sum_depos = dKZ+dUSD+dEUR
        
    ti9 = data_fi.loc[i,"transfers_amount_in_9"]
    ti10 = data_fi.loc[i,"transfers_amount_in_10"]
    to9 = data_fi.loc[i,"transfers_amount_out_9"]
    to10 = data_fi.loc[i,"transfers_amount_out_10"]
    tci9 = data_fi.loc[i,"transfers_count_in_9"]
    tci10 = data_fi.loc[i,"transfers_count_in_10"]
    tco9 = data_fi.loc[i,"transfers_count_out_9"]
    tco10 = data_fi.loc[i,"transfers_count_out_10"]
    
    if str(ti9)=="nan" or str(ti10)=="nan":
        ti = None
    else:
        ti = ti9+ti10
        
    if str(to9)=="nan" or str(to10)=="nan":
        to = None
    else:
        to = to9+to10

    if str(tci9)=="nan" or str(tci10)=="nan":
        tci = None
    else:
        tci = tci9+tci10
        
    if str(tco9)=="nan" or str(tco10)=="nan":
        tco = None
    else:
        tco = tco9+tco10
    
    
    difference=None
    if ti and to:
        difference = ti-to
    
    
    
    
    ac8 = data_fi.loc[i,"auth_count_8"]
    ac9 = data_fi.loc[i,"auth_count_9"]
    ac10 = data_fi.loc[i,"auth_count_10"]
    
    if str(ac8)=="nan" or str(ac9)=="nan" or str(ac10)=="nan":
        ac = None
    else:
        ac = ac8+ac9+ac10
    
    if sum_depos:
        frame.loc[i,"sum_depos"] = sum_depos
        
    if ti:
        frame.loc[i,"ti"] = ti
    
    if to:
        frame.loc[i,"to"] = to
        
    if tci:
        frame.loc[i,"tci"] = tci
        
    if tco:
        frame.loc[i,"tco"] = tco
        
    if ac:
        frame.loc[i,"ac"] = ac
    
    if difference:
        frame.loc[i,"dif"]=difference
        
        
```
